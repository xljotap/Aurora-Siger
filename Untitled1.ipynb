**Sistema de Verificação de Decolagem**

Este projeto simula a análise de telemetria de uma nave, verificando se está apta para decolagem com base em parâmetros de segurança.





## **1.1 Organização e descrição da telemetria**

**Regras de segurança**

- Temperatura interna: 18°C a 30°C
- Temperatura externa: -10°C a 45°C
- Integridade estrutural: deve ser 1
- Nível de energia: mínimo de 75%
- Pressão dos tanques: 60 a 110 unidades
- Status dos módulos: todos devem estar "OK"

## **1.2 Algoritmo de verificação**

## Dados analisados

- Temperatura interna
- Temperatura externa
- Integridade estrutural
- Nível de energia
- Pressão dos tanques
- Status dos módulos críticos

## **1.3 e 1.4 Algoritmo e Análise Energetica**

**Detalhamento do cálculo Energetico**

Cenário aprovado:
Temperatura interna: 25
Temperatura externa: 20
Integridade: 1
Energia: 80
Pressão: 90
modulos: OK

```python
def verificacao():

    print('Bem vindos ao algoritimo de segurança da Nave Orion')

    nome = input('Digite seu nome de astronauta: ')

    saudacao = input(f'Olá {nome}, está preparado para a verificação de segurança para a decolagem? (S/N) ').strip().lower()

    if saudacao == 's' or saudacao == 'sim':
        print('Ótimo! Vamos começar a verificação de segurança...\n')
    else:
        print('Missão abortada pelo usuário.')
        return

    temperatura_interna = float(input('Digite a temperatura interna da nave (em °C): '))
    temperatura_externa = float(input('Digite a temperatura externa da nave (em °C): '))
    integridade = int(input('Integridade estrutural (1 = ok / 0 = FALHA): '))
    energia = float(input('Nível de energia (em %): '))
    pressao = float(input('pressão dos tanques:'))
    modulos = input('Status dos módulos (OK/FALHA): ').strip().upper()

    if integridade != 1:
        print('Alerta: Integridade estrutural comprometida! Missão abortada.')
        return

    if energia < 75:
        print('Alerta: Nível de energia insuficiente! Missão abortada.')
        return

    if not (18 <= temperatura_interna <= 30):
        print('Alerta: Temperatura interna fora do intervalo seguro! Missão abortada.')
        return

    if not (-10 <= temperatura_externa <= 45):
        print('Alerta: Temperatura externa fora do intervalo seguro! Missão abortada.')
        return

    if not (60 <= pressao <= 110):
        print('Alerta: Pressão dos tanques fora do intervalo seguro! Missão abortada.')
        return

    if modulos != 'OK':
        print('Alerta: Status dos módulos comprometido! Missão abortada.')
        return

    print('\nCalculando autonomia energética...')

    capacidade_total = float(input('Digite a capacidade total da bateria (em kWh): '))
    carga_atual = energia
    consumo = float(input('Consumo na decolagem (kWh): '))
    perdas = float(input('Perdas energeticas (%): ')) / 100

    energia_disponivel = capacidade_total * (carga_atual / 100)
    energia_real = energia_disponivel * (1 - perdas)
    autonomia = energia_real / consumo

    horas = int(autonomia)
    minutos = int((autonomia - horas) * 60)

    print(f'Autonomia estimada: {horas} horas e {minutos} minutos.')

    print('\nRESULTADO FINAL:')
    print('PRONTO PARA DECOLAR')

verificacao()
```

# **1.5 ANÁLISE ASSISTIDA POR IA**

**Classificação dos dados**

- Os dados coletados foram classificados com base nos limites de segurança definidos no sistema, como temperatura interna entre 18°C e 30°C, temperatura externa entre -10°C e 45°C, energia acima de 75%, pressão entre 60 e 110, integridade estrutural igual a 1 e módulos com status "OK". Essa organização permite identificar rapidamente se as condições estão adequadas para a decolagem, separando situações seguras de possíveis riscos.

**Identificação de possíveis anomalias**

- As anomalias são identificadas quando algum dos dados sai dos padrões estabelecidos, como nível de energia abaixo de 75%, temperaturas fora do intervalo seguro, falha na integridade estrutural, pressão inadequada ou módulos com falha. Essas situações representam riscos diretos para a missão e podem levar ao cancelamento da decolagem, sendo essencial detectá-las antes da execução.

**Sugestões de risco**

- Para aumentar a segurança, recomenda-se implementar sistemas de alerta automático para condições críticas, como baixa energia ou pressão fora do padrão, além de monitoramento contínuo dos dados. Também é importante utilizar redundância em sistemas essenciais e realizar verificações automatizadas antes da decolagem, reduzindo a chance de falhas e aumentando a confiabilidade da operação.
